```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
SERVO_MIN_PULSEWIDTH_US=500 
SERVO_MAX_PULSEWIDTH_US=2500  
SERVO_MIN_DEGREE=-90   
SERVO_MAX_DEGREE=90    

SERVO_PULSE_GPIO=0        
SERVO_TIMEBASE_RESOLUTION_HZ=1000000 
SERVO_TIMEBASE_PERIOD=20000    

def example_angle_to_compare(angle):
    return (angle - SERVO_MIN_DEGREE) * (SERVO_MAX_PULSEWIDTH_US - SERVO_MIN_PULSEWIDTH_US) / (SERVO_MAX_DEGREE - SERVO_MIN_DEGREE) + SERVO_MIN_PULSEWIDTH_US;
```

```python
angle = np.arange(-90,90,1)

result = example_angle_to_compare(angle)

plt.plot(result)
plt.show()
```
